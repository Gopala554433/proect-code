```python
# To standardize the feature values, initialize an object called StandardScaler.
scaler = StandardScaler()

#Fit the data X to the StandardScaler, determining the standard deviation and mean for scaling
scaler.fit(X)

# To normalize the data (mean=0, std=1), transform it using the fitted scaler.
scaler_X = scaler.transform(X)
```

#Split Dataset  Train Test

```python
# data solitting into trainandtestset in the ratio 80%:20%, where 80% data usedfor trainsetmodel and 20% data usedfor testingsetmodel
Xtrain,Xtest,Ytrain,Ytest=train_test_split(scaler_X, Y, test_size=0.2, random_state=100)
```

```python
# checking the shape of the Xtrain-and-Xtest
Xtrain.shape, Xtest.shape
```

```python
# checking the shape of the Ytrain-and-Ytest
Ytrain.shape, Ytest.shape
```

#VotingClassifier

```python
# Initialize a DecisionTreeClassifier with a maximum depth of 20 and a fixed random state for reproducibility
clf1 = DecisionTreeClassifier(max_depth=20, random_state=0)

# Initialize an ExtraTreesClassifier with 10 trees (estimators), a maximum depth of 29, and a fixed random state
clf2 = ExtraTreesClassifier(n_estimators=10, max_depth=29, random_state=0)

# Initialize an AdaBoostClassifier with 10 weak learners (estimators)
clf3 = AdaBoostClassifier(n_estimators=10)

# Initialize a HistGradientBoostingClassifier with categorical cross-entropy as the loss function and a maximum depth of 10
clf4 = HistGradientBoostingClassifier(loss='categorical_crossentropy', max_depth=10, random_state=0)

# Create a VotingClassifier that combines the four classifiers, using soft voting for probabilistic prediction averaging
Voting_clf = VotingClassifier(estimators=[('DC', clf1), ('EXT', clf2), ('ADB', clf3), ('HGB', clf4)], voting='soft')
```

```python
# Import the warnings library to managewarning messages
import warnings

# Suppress warnings that may appear during model trainingset and predictionset
warnings.filterwarnings('ignore')

# Fit the VotingClassifier on the trainingdata(Xtrainset and Ytrainset)
Voting_clf.fit(Xtrain, Ytrain)

# Makepredictions on the testdata(Xtest) using the fitted VotingClassifier
pred_ytest = Voting_clf.predict(Xtest)
```

#Testing Accuracy

```python
# Assign an 8 by 5 inch figure size to the layout.
Mat_plot_lib.figure(figsize=[8,5])

# Determine which true labels (Ytest) and which predicted labels (pred_ytest) to use to calculate the confusion matrix.
cm = confusion_matrix(Ytest, pred_ytest)

# Use a heatmap, annotations, and a "cool" colormap to plot the confusion matrix.
# Set 'fmt' to 'g' to show the values of the confusion matrix as integers; 'linewidths' adds cell separation.
sns.heatmap(cm, annot=True, cmap="cool", fmt='g', linewidths=5)

# Increase the heatmap's title by adding one and changing the text size and vertical alignment.
Mat_plot_lib.title("CONFUSION MATRIX", y=1.1, fontsize=20)

# Show the ploted of the confusedmatrix
Mat_plot_lib.show()
```


#CLASSIFICATION REPORT

```python
# Printed the classificationreport, which includesprecisionscore, recallscore, f1score, and support for each class
print(classification_report(Ytest, pred_ytest))
```

```python
# Bring in the scikit-learn confusion_matrix function.
from sklearn.metrics import confusion_matrix

#From the genuine labels(Ytest) and the predictedlabels(pred_ytest), created the confusionmatrix.
conf_matrix = confusion_matrix(Ytest, pred_ytest)

# ExtractTruePositives(TP), FalsePositives(FP), TrueNegatives(TN), and FalseNegatives(FN)from-the-confusionmatrix
TP = conf_matrix[1, 1]  # TruePositives: correctlypredicted positivecases
FP = conf_matrix[0, 1]  # FalsePositives: incorrectlypredicted as positive
TN = conf_matrix[0, 0]  # TrueNegatives: correctlypredicted negativecases
FN = conf_matrix[1, 0]  # FalseNegatives: incorrectlypredicted as negative

# Establish gain values for various results.
gain_per_positive = 1              # Gained for each TruePositive
gain_per_false_positive = 5        # Gained for each FalsePositive
gain_per_true_negative = 2         # Gained for each TrueNegative
gain_per_false_negative = 5        # Gained for each FalseNegative

# Calculate the total gain for each outcome based on the confusion matrix and defined gain values
gain_TP = TP * gain_per_positive  # Gained from TruePositives
gain_FP = FP * gain_per_false_positive  # Gained from FalsePositives
gain_TN = TN * gain_per_true_negative  # Gained from TrueNegatives
gain_FN = FN * gain_per_false_negative  # Gained from FalseNegatives

# Add up all of the individual gains to determine the overall gain.
total_gain = gain_TP + gain_FP + gain_TN + gain_FN

# Output the overallgain computedusing the gain values and confusionmatrix.
print("Gain:", total_gain)
```
